### word2vecの引数
- sentences : 対象となる分かち書きされているテキスト
- size      : 出力するベクトルの次元数
- min_count : この数値よりも登場回数が少ない単語は無視する
- window    : 一つの単語に対してこの数値分だけ前後をチェックする
- sentences = word2vec.Text8Corpus('text8.txt')

```python
from gensim.models import word2vec
import logging
import sys

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
def train(file_name, output_name, size=200, min_count=1, window=5):
    input_file = file_name + '.txt'
    ouput_file = output_name+ '.model'
    sentences = word2vec.Text8Corpus(input_file)
    model = word2vec.Word2Vec(sentences, size=size, min_count=min_count, window=window)
    model.save(ouput_file)
    print("Finish!!!")
```

```python
def get_similar(model_file, posi, nega, n=10):
    model = word2vec.Word2Vec.load(model_file)
    result = model.most_similar(positive = posi, negative = nega, topn = n)
    for i, r in enumerate(result):
        print(i,'　', r[0],'　', r[1])
```
